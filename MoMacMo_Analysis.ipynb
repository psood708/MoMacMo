```python

import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt


file_path = 'sample_data/final_csv.csv'
df = pd.read_csv(file_path)

selected_features = ['x', 'y', 'volume', 'count', 'power','frame']
# selected_features1 = ['volume', 'count', 'power']


# It's important to standardize the data before applying PCA
scaler = StandardScaler()
df_standardized = scaler.fit_transform(df[selected_features])
```

```python
df_standardized
```

```python
n_components = 4
pca = PCA(n_components=n_components)

# Fit and transform the standardized data
principal_components = pca.fit_transform(df_standardized)

#  Explore the explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_
print("Explained Variance Ratio:", explained_variance_ratio)
print("Total Explained Variance:", sum(explained_variance_ratio))
```

```python
principal_components
```

```python
plt.bar(range(1, n_components + 1), explained_variance_ratio, alpha=0.5, align='center')
plt.title('Explained Variance Ratio for Each Principal Component')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.show()
```

```python
cumulative_explained_variance = explained_variance_ratio.cumsum()

# Determine the number of components to retain (e.g., 95% of the variance)
desired_variance_ratio = 0.80
selected_components = np.argmax(cumulative_explained_variance >= desired_variance_ratio) + 1

print(f"Number of components to retain for {desired_variance_ratio * 100}% variance: {selected_components}")

```

```python
# If you decide to keep, for example, 3 principal components
selected_components = 3
final_data = pca.transform(df_standardized)[:, :selected_components]
```

```python
final_data
```

```python
import matplotlib.pyplot as plt

# Assuming final_data has 2 or 3 columns (2D or 3D scatter plot)
if final_data.shape[1] == 2:
    plt.scatter(final_data[:, 0], final_data[:, 1])
    plt.title('PCA: 2D Scatter Plot')
    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    plt.show()
elif final_data.shape[1] == 3:
    from mpl_toolkits.mplot3d import Axes3D

    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(final_data[:, 0], final_data[:, 1], final_data[:, 2])
    ax.set_title('PCA: 3D Scatter Plot')
    ax.set_xlabel('Principal Component 1')
    ax.set_ylabel('Principal Component 2')
    ax.set_zlabel('Principal Component 3')
    plt.show()
else:
    print("Can only plot 2D or 3D scatter plots.")
```

```python
import plotly.express as px
import numpy as np
from sklearn.cluster import DBSCAN

# Step 1: Instantiate the DBSCAN model
dbscan_model = DBSCAN(eps=0.4, min_samples=3)  # Adjust parameters as needed

# Step 2: Fit the model on the final_data
labels = dbscan_model.fit_predict(final_data)

# Step 3: Visualize the clusters
if final_data.shape[1] == 2:
    fig = px.scatter(x=final_data[:, 0], y=final_data[:, 1], color=labels, title='DBSCAN Clustering: 2D Scatter Plot',
                     labels={'color': 'Cluster'}, color_continuous_scale='plasma')
    fig.update_layout(xaxis_title='Principal Component 1', yaxis_title='Principal Component 2')
    fig.show()
elif final_data.shape[1] == 3:
    fig = px.scatter_3d(x=final_data[:, 0], y=final_data[:, 1], z=final_data[:, 2], color=labels,
                        title='DBSCAN Clustering: 3D Scatter Plot', labels={'color': 'Cluster'}, color_continuous_scale='viridis')
    fig.update_layout(scene=dict(xaxis_title='Principal Component 1', yaxis_title='Principal Component 2', zaxis_title='Principal Component 3'))
    fig.show()
else:
    print("Can only visualize 2D or 3D scatter plots.")

# Step 4: Examine the cluster labels
print("Cluster Labels:", labels)
```

```python
#determining K for our analysis
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Step 1: Define a range of k values
k_values = range(1, 11)  # You can adjust this range based on your problem

# Step 2: Fit KMeans models for each k and compute WCSS
wcss = []
for k in k_values:
    kmeans_model = KMeans(n_clusters=k, random_state=42)
    kmeans_model.fit(final_data)
    wcss.append(kmeans_model.inertia_)

# Step 3: Plot the Elbow curve
plt.plot(k_values, wcss, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.show()
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Assuming df is your DataFrame and it's already loaded
df = pd.read_csv('./sample_data/test.csv')

# Selecting the 'x', 'y', and 'z' columns
selected_columns = df[['x', 'y', 'z']]

# Standardizing the features
x = StandardScaler().fit_transform(selected_columns)

# Applying PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)

# Creating a DataFrame with the principal components
df_pca = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2'])

print(df_pca)
```

```python
from sklearn.cluster import DBSCAN

# Assuming df_pca is your DataFrame with the principal components
X = df_pca.values

# Define the DBSCAN model
dbscan = DBSCAN(eps=0.5, min_samples=5)

# Fit the model to your data
dbscan.fit(X)

# Get the cluster labels (assigning the label -1 for noise)
labels = dbscan.labels_

# Add the cluster labels as a new column in your DataFrame
df_pca['Cluster'] = labels

print(df_pca)
```

```python
import matplotlib.pyplot as plt

# Create a scatter plot of the two principal components
plt.figure(figsize=(10,10))
scatter = plt.scatter(df_pca['principal component 1'], df_pca['principal component 2'], c=df_pca['Cluster'], cmap='viridis')

# Add a color bar and labels
plt.colorbar(scatter)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('DBSCAN Clustering (PCA Reduced Data)')

# Show the plot
plt.show()
```
